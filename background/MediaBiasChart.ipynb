## Media Bias Chart - V4.0 

```python
import pandas as pd
import numpy as np
```

```python
# from https://www.adfontesmedia.com/
media_bias = pd.read_csv("MediaBiasChart Version4.0.csv")
```

```python
quality_threshold = 24
bias_threshold = 18
```

```python
mask = (media_bias["Vertical Rank"] > quality_threshold) & \
       (np.abs(media_bias["Horizontal Rank"]) < bias_threshold)
```

```python
selected_media = media_bias[mask]
```

```python
conservative_media = selected_media[selected_media["Horizontal Rank"] > 0]
liberal_media = selected_media[selected_media["Horizontal Rank"] < 0]
```
