# Embeddings Visualization

```python
import pickle
import os
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
```

```python
MODELS_FOLDER = "../../models/word_embeddings"
```

```python
models = {
    1: {
        "name": "Webhose & Google News - All articles",
        "base_name": "wb_gn_full"
    },
    2: {
        "name": "Webhose & Google News - Only liberal sources",
        "base_name": "wb_gn_liberal"
    },
    3: {
        "name": "Webhose & Google News - Only conservatives sources",
        "base_name": "wb_gn_conservative"
    },
    4: {
        "name": "All the news - All articles",
        "base_name": "atn_full"
    },
    5: {
        "name": "Webhose & Google News - Only liberal sources",
        "base_name": "atn_liberal"
    },
    6: {
        "name": "Webhose & Google News - Only conservatives sources",
        "base_name": "atn_conservative"
    },    
}
```

```python
def load_data(filename):
    with open(filename, 'rb') as pickle_file:
        return pickle.load(pickle_file)
```

```python
def reduce_with_tSNE(embeddings, reversed_dictionary, num_words=500):
    tsne = TSNE(perplexity=30, n_components=2, init='pca', n_iter=5000)
    num_words = 500
    low_dim_embeds = tsne.fit_transform(embeddings[:num_words, :])
    labels = [reversed_dictionary[i] for i in range(num_words)]
    
    return low_dim_embeds, labels
```

```python
def load_and_reduce_dim(base_name, num_words):
    input_filename = os.path.join(MODELS_FOLDER, "{}_embedding_vocab.pickle".format(base_name))
    input_data = load_data(input_filename)
    embeddings = input_data["embeddings"]
    reversed_dictionary = input_data["reversed_dictionary"]
    
    return reduce_with_tSNE(embeddings, reversed_dictionary, num_words)
```

```python
def plot_with_labels(low_dim_embeds, labels, filename):
    plt.figure(figsize=(18, 18))  # in inches
    for i, label in enumerate(labels):
        x, y = low_dim_embeds[i, :]
        plt.scatter(x, y)
        plt.annotate(label,
                     xy=(x, y),
                     xytext=(5, 2),
                     textcoords='offset points',
                     ha='right',
                     va='bottom')

    plt.savefig(filename)
    plt.show()
```

```python
selected_model = 1
base_name = models[selected_model]["base_name"]
models[selected_model]["name"]
```

```python
embeds, labels = load_and_reduce_dim(base_name, 500)
```

```python
plot_name = "{}_tsne.png".format(base_name)
plot_with_labels(embeds, labels, plot_name)
```
